```python
full_list = [scan_simple, scan_length, scan_add_turn_left, cfq_mcd1, cogs_test, cogs_lex, cogs_stru, smcal, pcfgset]
full_list_str = ['scan_simple', 'scan_length', 'scan_add_turn_left', 'cfq_mcd1', 'cogs_test', 'cogs_lex', 'cogs_stru', 'smcal', 'pcfgset']

for x in range(len(full_list)-1): 
  for y in range(x+1, len(full_list)):
    coff_score = np.corrcoef(full_list[x], full_list[y])[0][1]
    # print(full_list_str[x], full_list_str[y], round(coff_score, 3))
    plt.figure(figsize=(8,6))
    sns.regplot(full_list[x], full_list[y])
    plt.xlabel(f'{full_list_str[x]}', fontsize=24)
    plt.ylabel(f'{full_list_str[y]}', fontsize=24)
    plt.title('Correlation Corrcoef: {:.3f}'.format(coff_score), fontsize=24)
    plt.show()
    print()
```
